```python
import urllib.request
import bs4
import json
import pickle
import sys

import sys
sys.setrecursionlimit(10000)

# 検索対象url番号
start, end = 0, 10#int(sys.argv[1]), int(sys.argv[2])

# 取得した数
num = 0
menu = []

for i in range(start, end):
    tmp = i+1 # 次に検索し始めるurl番号を記録
    url = 'https://www.asken.jp/calculate/meal/' + str(i)
    try:
        
        me = bs4.BeautifulSoup(urllib.request.urlopen(url).read(), "lxml")
        num += 1
        print("found:", url, ", scraper:", str(start) + '-' + str(end), ", url_id:", i, ", num:" , num)
        me = me.find('div', id="content_all_body")
        menu.append(me)
        
    except urllib.error.HTTPError as e:
        print("[W] URL Not Found")
        continue

with open('data/menu-' + '{0:07d}'.format(start)+ '-' + '{0:07d}'.format(end) + '.pkl', 'wb') as f:
     pickle.dump(menu, f)
```

```python
import os

path_data = os.listdir('data/')

with open('data/' + path_data[0], 'rb') as f:
     menu = pickle.load(f)

print(len(menu))

for me in menu:
    title = me.find("h1")
    print(title)
```
